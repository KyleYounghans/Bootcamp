```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "Pyber_CSV_Data/Pyber_City_Data.csv"
ride_data_to_load = "Pyber_CSV_Data/Pyber_Ride_Data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset
pyber_df = pd.merge(ride_data, city_data, on="city", how="left")
# Display the data table for preview
pyber_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

#Urban
urban_city_df = pyber_df[pyber_df["type"] == "Urban"].groupby([pyber_df["city"]])
urban_ID = urban_city_df["ride_id"].count()
urban_fare = urban_city_df["fare"].mean()
urban_count = urban_city_df["driver_count"].mean()

#Suburban
suburban_city_df = pyber_df[pyber_df["type"] == "Suburban"].groupby([pyber_df["city"]])
suburban_ID = suburban_city_df["ride_id"].count()
suburban_fare = suburban_city_df["fare"].mean()
suburban_count = suburban_city_df["driver_count"].mean()

#Rural
rural_city_df = pyber_df[pyber_df["type"] == "Rural"].groupby([pyber_df["city"]])
rural_ID = rural_city_df["ride_id"].count()
rural_fare = rural_city_df["fare"].mean()
rural_count = rural_city_df["driver_count"].mean()

# Build the scatter plots for each city types
plt.scatter(urban_ID, urban_fare, label = "Urban", s=urban_count*5, color=["coral"], edgecolor="black", alpha = 0.75, marker="o")
plt.scatter(suburban_ID, suburban_fare, label = "Suburban", s=suburban_count*5, color=["skyblue"], edgecolor="black", alpha = 0.75, marker="o")
plt.scatter(rural_ID, rural_fare, label = "Rural", s=rural_count*5, color=["gold"], edgecolor="black", alpha = 0.75, marker="o")

# Incorporate the other graph properties
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Pyber Ride Sharing Data (2016)")

# Create a legend
legend = plt.legend(fontsize=10, title= "City Types", loc="best")
legend.legendHandles[0]._sizes = [50]
legend.legendHandles[1]._sizes = [50]
legend.legendHandles[2]._sizes = [50]

# Incorporate a text label regarding circle size
plt.text(42,35,"Note: \nCircle size correlates with driver count per city.", fontsize = 10)

# Save Figure
plt.savefig("Pyber_Images/PyberRideSharingData.png")

#Show Chart
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
group_df = pyber_df.groupby(['type'])
fare_df = group_df['fare'].sum()

# Build Pie Chart
labels = ["Rural","Suburban","Urban"]
explode = (0, 0, 0.125)
colors = ["gold", "skyblue", "coral"]
plt.pie(fare_df, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=135)
plt.title("% of Total Fares by City Type")

# Save Figure
plt.savefig("Pyber_Images/Percent_of_Total_Fares_by_City_Type.png")

#Show Chart
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
rides_df = group_df['ride_id'].count()

# Build Pie Chart
labels = ["Rural","Suburban","Urban"]
explode = (0, 0, 0.125)
colors = ["gold", "skyblue", "coral"]
plt.pie(rides_df, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=135)
plt.title("% of Total Rides by City Type")

# Save Figure
plt.savefig("Pyber_Images/Percent_of_Total_Rides_by_City_Type.png")

# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
drivers_group_df = city_data.groupby(['type'])
drivers_df = drivers_group_df['driver_count'].sum()

# Build Pie Charts
labels = ["Rural","Suburban","Urban"]
explode = (0, 0, 0.125)
colors = ["gold", "skyblue", "coral"]
plt.pie(drivers_df, explode=explode, labels=labels, colors=colors,autopct="%1.1f%%", shadow=True, startangle=135)
plt.title("% of Total Drivers by City Type")

# Save Figure
plt.savefig("Pyber_Images/Percent_of_Total_Drivers_by_City_Type.png")

# Show Figure
plt.show()
```
